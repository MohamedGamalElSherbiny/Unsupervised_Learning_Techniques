```python
from plotnine import *
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = pd.Series(data.target)
```

```python
print(df.columns)
df.head()
df.shape
```

```python
features = df.columns[0:30] #grab all numeric columns of interest
z = StandardScaler()
df[features] = z.fit_transform(df[features])
```

```python
ggplot(df, aes(x = "mean radius", y = "mean texture")) + geom_point()
```

```python
pca = PCA()
pca.fit(df[features])
pca.explained_variance_ratio_
```

```python
pcaDF = pd.DataFrame({"pc": range(1,31),"expl_var" : pca.explained_variance_ratio_,
                      "cum_var": pca.explained_variance_ratio_.cumsum()})
pcaDF.head()
```

```python
(ggplot(pcaDF, aes(x = "pc", y = "expl_var")) + geom_line() + geom_point())
```

```python
(ggplot(pcaDF, aes(x = "pc", y = "cum_var")) + geom_line(color = "pink") +
 geom_point(color = "pink") + geom_hline(yintercept = 0.95))
```

```python
data = pca.transform(df[features])
df4 = pd.DataFrame(data[:,0:4])


df10 = pd.DataFrame(data[:, 0:10])

#modeMod1
lr1 = LogisticRegression()
lr1.fit(df[features], df["target"])
print("all data: ", lr1.score(df[features], df["target"]))

#modeMod1
lr2 = LogisticRegression()
lr2.fit(df10, df["target"])
print("10 PCs:   ", lr2.score(df10, df["target"]))

#modeMod1
lr3 = LogisticRegression()
lr3.fit(df4, df["target"])
print("4 PCs:    ", lr3.score(df4, df["target"]))
```

## Iris Dataset:

```python
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = pd.Series(data.target)
```

```python
df.columns
```

```python
df.shape
```

```python
labels = ['setosa', 'versicolor', 'virginica']
title = 'Sepal Width vs Sepal Length Iris Dataset'
```

```python
g = sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=df,
                    palette='colorblind')
g.set(title=title)
plt.xticks(rotation=90)
legend_labels, _= g.get_legend_handles_labels()
plt.legend(legend_labels, labels)
plt.show()
```

```python
X = df.drop(['target'], axis='columns')
y = df['target']
```

```python
X.head()
```

```python
features = X.columns
```

```python
standard_scaler = StandardScaler()
X[features] = standard_scaler.fit_transform(X)
```

```python
pca = PCA()
pca.fit(X)
pca.explained_variance_ratio_
```

```python
pcaDF = pd.DataFrame({"pc": range(0,4),"expl_var" : pca.explained_variance_ratio_,
                      "cum_var": pca.explained_variance_ratio_.cumsum()})
pcaDF.head()
```

```python
(ggplot(pcaDF, aes(x = "pc", y = "expl_var")) + geom_line() + geom_point())
```

```python
(ggplot(pcaDF, aes(x = "pc", y = "cum_var")) + geom_line(color = "pink") +
 geom_point(color = "pink") + geom_hline(yintercept = 0.95))
```

```python
X.head()
```

```python
y.head()
```

```python
data = pca.transform(X)
X1 = pd.DataFrame(data[:,0])

X2 = pd.DataFrame(data[:, 0:2])

#modeMod1
lr1 = LogisticRegression()
lr1.fit(X, y)
print("all data: ", lr1.score(X, y))

#modeMod1
lr2 = LogisticRegression()
lr2.fit(X1, y)
print("1 PCs:   ", lr2.score(X1, y))

#modeMod1
lr3 = LogisticRegression()
lr3.fit(X2, df["target"])
print("2 PCs:    ", lr3.score(X2, y))
```
